```python
import pandas as pd
```

```python

class PrintColors:
    """
    **** change the colors used to print
    Example of use:
    print(f'{PrintColors.BOLD} this is printed in bold  {PrintColors.RESET} comm back to normal
    {PrintColors.foreground_rgb(100,100,255)} blue fore{PrintColors.background_rgb(200,0,100)}
    changed background {PrintColors.RESET}')
    """
    RESET = '\033[0m'
    BOLD = '\033[1m'
    ITALIC = '\033[3m'
    UNDERLINE_ON = '\033[4m'
    UNDERLINE_OFF = '\033[24m'
    BLINK_SLOW = '\033[5m'
    BLINK_FAST = '\033[6m'
    BLINK_OFF = '\033[25m'
    OKBLUE = '\033[94m'
    OKGREEN = '\033[92m'
    WARNING = '\033[93m'
    FAIL = '\033[91m'
    # standard colors
    RED = '\033[38;2;255;0;0m'
    BACKGROUND_RED = '\033[48;2;255;0;0m'
    LIME = '\033[38;2;0;255;0m'
    BACKGROUND_LIME = '\033[48;2;0;255;0m'
    BLUE = '\033[38;2;0;0;255m'
    BACKGROUND_BLUE = '\033[48;2;0;0;255m'
    GREEN = '\033[38;2;0;128;0m'
    BACKGROUND_GREEN = '\033[48;2;0;128;0m'
    YELLOW = '\033[38;2;255;255;0m'
    BACKGROUND_YELLOW = '\033[48;2;255;255;0m'
    MAGENTA = '\033[38;2;255;0;255m'
    BACKGROUND_MAGENTA = '\033[48;2;255;0;255m'
    CYAN = '\033[38;2;0;255;255m'
    BACKGROUND_CYAN = '\033[48;2;0;255;255m'
    GRAY = '\033[38;2;128;128;128m'
    BACKGROUND_GRAY = '\033[48;2;128;128;128m'
    ORANGE = '\033[38;2;255;94;19m'
    BACKGROUND_ORANGE = '\033[48;2;255;94;19m'
    BROWN = '\033[38;2;136;69;19m'
    BACKGROUND_BROWN = '\033[48;2;136;69;19m'

    @staticmethod
    def foreground_rgb(r, g, b):
        return f'\033[38;2;{r};{g};{b}m'

    @staticmethod
    def background_rgb(r, g, b):
        return f'\033[48;2;{r};{g};{b}m'

    @staticmethod
    def set_foreground(r, g, b):
        print(PrintColors.foreground_rgb(r, g, b))

    @staticmethod
    def set_background(r, g, b):
        print(PrintColors.background_rgb(r, g, b))

```
